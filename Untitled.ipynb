```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
X_df = pd.DataFrame(np.random.randint(0, 10, size=(100, 100)))
```

```python
X_df
```

```python
X_df.iloc[np.random.randint(0, 100, 10), 1] = np.nan
```

```python
from sklearn.pipeline import Pipeline
p = Pipeline([
    ('imputer', SimpleImputer(strategy="constant", fill_value=-9999)),
    ('hot', OneHotEncoder()), 
])
```

```python
transformers = [("aman", p, [0,1,2,3,4])]
```

```python
transformers
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(transformers)
```

```python
p1 = Pipeline([
    ('abc', preprocessor)
])
```

```python
pipe_cat = "C:\\Users\\amand\\Desktop\\" + "abc.plk"
```
